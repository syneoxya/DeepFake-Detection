## Sampling frames and storing them in a separate database

### Importing libraries

```python
import cv2
import os
import random
import numpy as np
from mtcnn.mtcnn import MTCNN
import dlib
from IPython.display import Image, display
import matplotlib.pyplot as plt
```

### Function to sample the frames and save the videos

```python
def create_empty_video(video_file_path, fps, frame_size):
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(video_file_path, fourcc, 10, frame_size, isColor=False)
    out.release()
def sample_and_save_video(video_file, output_folder, num_frames):
    cap = cv2.VideoCapture(video_file)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    if (total_frames < num_frames):
        return
    # getting sampled frame indices
    sampled_frame_indices = random.sample(range(total_frames), num_frames)
    frames_to_save = []

    current_frame_index = 0
    frame_count = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        if current_frame_index in sampled_frame_indices:
            # Preprocess frame as needed
            # ...
            preprocessed_frame = frame
            frames_to_save.append(preprocessed_frame)

        current_frame_index += 1

    cap.release()
    # Saving sampled frames as a new video
    output_video_filename = os.path.basename(video_file)
    output_video_path = os.path.join(output_folder, f"samples_{output_video_filename}")
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    # opening the output file
    create_empty_video(output_video_path, cap.get(cv2.CAP_PROP_FPS), (frames_to_save[0].shape[1], frames_to_save[0].shape[0]))

    out = cv2.VideoWriter(output_video_path, fourcc, 10.0, (frames_to_save[0].shape[1], frames_to_save[0].shape[0]))
    if not out.isOpened():
        print(f"Error: Could not open video writer for {output_video_path}. Skipping.")
        return
    
    for frame_to_save in frames_to_save:
        out.write(frame_to_save)

    out.release()
```

### Sampling videos

```python
def process_videos(input_folder, output_folder, num_frames_per_video):
    # Iterating through videos
    os.makedirs(output_folder, exist_ok=True)
    videos = os.listdir(input_folder)
    ctr = 1
    # Iterate through videos in the class folder
    for video in videos:
        video_file = os.path.join(input_folder, video)
        # Sample frames from the video and save them as a new video
        sample_and_save_video(video_file, output_folder, num_frames=num_frames_per_video)
        print("video ", ctr, " done")
        ctr +=1
```

### Extracting the faces

```python
def extract_faces(input_video_path, output_folder, margin_percentage=10, output_face_size=(100, 100)):
    os.makedirs(output_folder, exist_ok=True)

    # Loading MTCNN
    detector = MTCNN()

    # Loading video
    cap = cv2.VideoCapture(input_video_path)

    # Getting meta data
    fps = cap.get(cv2.CAP_PROP_FPS)
    frame_size = (int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)), int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))

    # adding a ten percent margin
    margin_x = int(frame_size[0] * margin_percentage / 100)
    margin_y = int(frame_size[1] * margin_percentage / 100)

    # Set the desired size for the extracted face regions
    output_face_size = (output_face_size[0] + 2 * margin_x, output_face_size[1] + 2 * margin_y)



    output_video_filename = os.path.basename(input_video_path)
    output_video_path = os.path.join(output_folder, f"samples_{output_video_filename}")
    if os.path.exists(output_video_path):
        print(f"Video file '{output_video_path}' already exists. Skipping processing.")
        return

    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_video_path, fourcc, fps, output_face_size)

    while True:
        ret, frame = cap.read()

        if not ret:
            break

        # finding face
        faces = detector.detect_faces(frame)

        # getting bounding box
        for face in faces:
            x, y, width, height = face['box']

            # Adding the margin
            x -= margin_x
            y -= margin_y
            width += 2 * margin_x
            height += 2 * margin_y

            x = max(0, x)
            y = max(0, y)
            width = min(frame_size[0] - 1, width)
            height = min(frame_size[1] - 1, height)

            # Extracting the face region with margin
            face_region = frame[y:y+height, x:x+width]

            # Resize the face region to the desired size
            face_region_resized = cv2.resize(face_region, output_face_size)

            # Write the resized face region to the output video
            out.write(face_region_resized)

    # Release the VideoCapture and VideoWriter objects
    cap.release()
    out.release()

    cv2.destroyAllWindows()
```

#### Running it for all sampled videos

```python
def extract_all_faces(input_folder, output_folder):
    # Iterating through videos
    os.makedirs(output_folder, exist_ok=True)
    videos = os.listdir(input_folder)
    ctr = 1
    # Iterate through videos in the class folder
    for video in videos:
        video_file = os.path.join(input_folder, video)
        # Sample frames from the video and save them as a new video
        extract_faces(video_file, output_folder)
        print("video ", ctr, " done")
        ctr +=1
        if(ctr == 889*2+1):
            return
```
